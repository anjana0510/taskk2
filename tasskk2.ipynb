```python
# Step 1: Install necessary libraries
# !pip install diffusers transformers accelerate torch

# Step 2: Import libraries
import torch
from diffusers import StableDiffusionPipeline

# Step 3: Load the model
# You might need to accept the license for certain models on Hugging Face Hub.
# For example, to use 'runwayml/stable-diffusion-v1-5', you need to accept its license.
# Make sure you are logged in to Hugging Face if required: huggingface-cli login
model_id = "runwayml/stable-diffusion-v1-5" # Or any other Stable Diffusion model
try:
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32)
    if torch.cuda.is_available():
        pipe = pipe.to("cuda") # Move the model to GPU if available
except Exception as e:
    print(f"Error loading model: {e}")
    print("Please ensure you have accepted the model license on Hugging Face Hub and are logged in if necessary.")
    print("You can log in using: huggingface-cli login")
    exit() # Exit if model cannot be loaded

# Step 4: Define Text Prompt
prompt = "A futuristic city with flying cars and neon lights, highly detailed, cyberpunk style"

# Step 5: Generate Image
print(f"Generating image for prompt: '{prompt}'...")
try:
    image = pipe(prompt).images[0]
    print("Image generated successfully!")
except Exception as e:
    print(f"Error during image generation: {e}")
    exit()

# Step 6: Display/Save Image
# You'll need PIL (Pillow) to display/save the image
# !pip install Pillow
from PIL import Image

# Display the image (if in an environment that supports it, e.g., Jupyter Notebook)
# image.show()

# Save the image
image.save("futuristic_city.png")
print("Image saved as 'futuristic_city.png'")
```
